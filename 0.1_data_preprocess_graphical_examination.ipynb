```python
from google.colab import drive
drive.mount('/content/drive')
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# read dataset.
df = pd.read_csv("/content/drive/My Drive/2023_fall/BA305/BA_305_Project/Process_data/weatherAUS.csv")
df.RainTomorrow.replace(('Yes', 'No'), (int(1), int(0)), inplace=True)
df.head(20)
```

```python
# Step one, Graphical Examination
df.info()
```

```python
missing_percentage = df.isnull().mean() * 100

# Plotting
plt.figure(figsize=(10, 6))
missing_percentage.plot.bar()
plt.ylabel('Percentage of Missing Values')
plt.title('Missing Values Percentage by Feature')
plt.xticks(rotation=45)
plt.show()
```

```python
num_variables = list(df._get_numeric_data().columns)
num_variables
```

```python
cat_variables = list(set(df.columns) - set(num_variables))
cat_variables.remove('Location')
cat_variables.remove('Date')
cat_variables
```

```python
for feature in num_variables:
  plt.figure(figsize=(10,5))
  sns.kdeplot(df[[feature]],fill=True)
  plt.xlabel(f'{feature} Value')
  plt.ylabel('Distribution')
  plt.title(f'{feature} Data Density Distribution')
  plt.show()
```

```python
# plot bar char for cat variables
for feature in cat_variables:
  plt.figure(figsize=(10,5))
  sns.countplot(x = feature, data = df)
  plt.xlabel(f'{feature} Categories')
  plt.ylabel('Count')
  plt.title(f'{feature} Categories Data Count')
  plt.show()
```

```python


plt.figure(figsize=(10,8))
sns.heatmap(df[num_variables].corr())
plt.title('Correlation Heatmap')
plt.show()
```
